```python
import numpy as np
import math 
import matplotlib.pyplot as plt
from ipywidgets import interact


dimx=1500
dimy=1000
angL =int( math.atan(dimy/dimx)*180/np.pi)


# La escala espacial en cada eje es 1/cos(newAngle)
def scale(angL):
    global scale
    scale=np.empty([360])
    torad=np.pi/180;
    for i in range(180):
        if(i<=angL):
            scale[i+180]=scale[i]=1/math.cos(i*torad);
        if(i>angL and i<=90):
            scale[i+180]=scale[i]=1/math.cos((90-i)*torad);
        if(i>90 and i<=180-angL):
            scale[i+180]=scale[i]=1/math.cos((i-90)*torad);
        if(i>=180-angL):
            scale[i+180]=scale[i]=1/math.cos((180-i)*torad);


scale(angL);

rho_1k = np.arange(1000)
rho_32 = np.arange(32)
rho_64 = np.arange(64)
rho_128 = np.arange(128)
theta = np.linspace(0,2*np.pi,360)
A1k, R1k = np.meshgrid(theta, rho_1k)
A32, R32 = np.meshgrid(theta, rho_32)
A64, R64 = np.meshgrid(theta, rho_64)
A128, R128 = np.meshgrid(theta, rho_128)


# Función para crear el gráfico. No usada
def create_polar_plot(data):
    plt.pcolormesh(data, cmap='jet')
    plt.colorbar()
    plt.show()
    
```

```python
wins=64
wins2=32
nimg1=np.zeros([1000,360])
nimg2=np.zeros([1000,360])
nimg3=np.zeros([1000,360])
ncep1=np.zeros([1000,360])
ncep2=np.zeros([1000,360])
ncep3=np.zeros([1000,360])
interpI=np.linspace(0, 1, num=wins)
interpC=np.linspace(0, 1, num=wins2)
for i in range(360):
    val=int(500*scale[i])
    interp=np.linspace(0, 1, num=val)
    resultI1=np.interp(interp,interpI,img1[:,i])
    resultI2=np.interp(interp,interpI,img2[:,i])
    resultI3=np.interp(interp,interpI,img3[:,i])
    resultC1=np.interp(interp,interpC,cep1[:,i])
    resultC2=np.interp(interp,interpC,cep2[:,i])
    resultC3=np.interp(interp,interpC,cep3[:,i])
    
    array_destino = np.zeros(1000)
    array_destino[:val] = resultI1[:val]
    nimg1[:,i]=array_destino

    array_destino = np.zeros(1000)
    array_destino[:val] = resultI2[:val]
    nimg2[:,i]=array_destino

    array_destino = np.zeros(1000)
    array_destino[:val] = resultI3[:val]
    nimg3[:,i]=array_destino

    array_destino = np.zeros(1000)
    array_destino[:val] = resultC1[:val]
    ncep1[:,i]=array_destino

    array_destino = np.zeros(1000)
    array_destino[:val] = resultC2[:val]
    ncep2[:,i]=array_destino

    array_destino = np.zeros(1000)
    array_destino[:val] = resultC3[:val]
    ncep3[:,i]=array_destino

```
